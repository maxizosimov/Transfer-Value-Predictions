# Pulling Understat Player Data
This notebook is used to pull understat data into the aggregated window format
necessary for use in training both the LSTM and Bayesian model.

```python
import pandas as pd
import os
import sys
from understatapi import UnderstatClient
# Also import local helpers
target_dir = os.path.abspath('..') 

if target_dir not in sys.path:
    sys.path.append(target_dir)
    
from preprocess import get_position_players_stats_df
```

```python
# Create understat connection
understat = UnderstatClient()
```

```python
# Choose what league(s), season(s), position(s), what stats to use, and window size
leagues = ["EPL", "La_Liga", "Bundesliga", "Serie_A", "Ligue_1"]
seasons = list(range(2014,2026))
positions = ['M']
# note there are more features than in the final model- because the stats
# we use are a subset of them we don't need to rerun
stats = ["xG", "xA", "key_passes", "xGChain", "xGBuildup"] 
per_90_stats = list(map(lambda s: f"{s}_per_90", stats))
games_per_block = 10
```

```python
# Call function to pull data

stats_df = get_position_players_stats_df(understat, positions, games_per_block, stats, leagues=leagues, seasons=seasons)

# save dataset to csv
os.makedirs("data", exist_ok=True)  
stats_df.to_csv(f"../data/{'_'.join(positions)}_stats.csv", index=True)
```
